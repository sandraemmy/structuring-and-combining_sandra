```python
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv')
display(data.head())
```

```python
display(data.columns)
```

```python
#data.pivot(columns="sales_channel", values="total_claim_amount")
pivot_table = data.pivot_table(values="total_claim_amount", index="sales_channel", aggfunc="sum").round(2)
pivot_table
```
